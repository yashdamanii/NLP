```python
import pandas as pd
import re

# Load the dataset containing input_text, target_gender, and expected_output
df = pd.read_csv("pronoun_testcases.csv")

# Function to swap gender pronouns in a sentence based on the target gender
def swap_gender_pronouns(text, target):
    # Add spaces around punctuation to separate them from words for proper tokenization
    text = re.sub(r'([.,!?])', r' \1 ', text)
    words = text.split()  # Tokenize the text by whitespace
    swapped = []  # Store the transformed words here

    for i, word in enumerate(words):
        lw = word.lower()  # Lowercased version for comparison
        next_word = words[i + 1].lower() if i + 1 < len(words) else ""  # Lookahead for 'her'

        # Pronoun replacements when target is female
        if target == "female":
            if lw == "he":
                replacement = "she"
            elif lw == "him":
                replacement = "her"
            elif lw == "his":
                replacement = "her"
            elif lw == "himself":
                replacement = "herself"
            else:
                replacement = word  # Leave unchanged

        # Pronoun replacements when target is male
        elif target == "male":
            if lw == "she":
                replacement = "he"
            elif lw == "herself":
                replacement = "himself"
            elif lw == "her":
                # Context check: if next word looks like a noun, treat as possessive → 'his'
                if next_word.isalpha() and next_word not in [
                    'is', 'was', 'has', 'have', 'can', 'will',
                    'would', 'should', 'could', 'had', 'did', 'does'
                ]:
                    replacement = "his"
                else:
                    replacement = "him"
            elif lw == "hers":
                replacement = "his"
            else:
                replacement = word  # Leave unchanged

        else:
            replacement = word  # Leave unchanged for unknown gender

        # Preserve original capitalization (e.g. "He" → "She")
        if word.istitle():
            replacement = replacement.capitalize()

        swapped.append(replacement)

    # Join the swapped words into a sentence and fix spacing around punctuation
    result = ' '.join(swapped)
    result = re.sub(r'\s+([.,!?])', r'\1', result)
    return result

# Apply the function row-by-row to generate the predicted output
df["predicted_output"] = df.apply(
    lambda row: swap_gender_pronouns(row["input_text"], row["target_gender"]),
    axis=1
)

# Set display option to show all rows of the DataFrame
pd.set_option("display.max_rows", None)

# Display all relevant columns to review results
print(df[["input_text", "target_gender", "expected_output", "predicted_output"]])
```
